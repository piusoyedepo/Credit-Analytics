```python
df_loan_data_basic_var = \
    cs_retail_loan_cust.merge(
    cs_retail_loan_cust_info, on=['BRA_CODE','CUS_NUM'], how='left').merge(
    cs_retail_loan_cust_has_biz_acc_info, on=['BRA_CODE','CUS_NUM'], how='left').merge(
    cs_retail_loan_cust_acc_info, on=['BRA_CODE','CUS_NUM'], how='left').merge( 
    cs_retail_loan_cust_turn_lodge_count_info, on=['BRA_CODE','CUS_NUM'], how='left').merge(
    cs_digital_trxns_info, on=['BRA_CODE','CUS_NUM'], how='left').merge(
    cs_betting_religious_visa_util, on=['BRA_CODE','CUS_NUM'], how='left').merge( 
    cs_pos_trxns_info, on=['BRA_CODE','CUS_NUM'], how='left').merge(
    cs_airtime_info, on=['BRA_CODE','CUS_NUM'], how='left').merge( 
    cs_trxn_ntwrk_info, on=['BRA_CODE','CUS_NUM'], how='left').merge( 
    cs_retail_loan_cust_in_branch_info, on=['BRA_CODE','CUS_NUM'], how='left').merge( 
    cs_trxn_hour_period, on=['BRA_CODE','CUS_NUM'], how='left')
    
print 'Number of rows ', len(df_loan_data_basic_var)
print 'Number of columns ', len(df_loan_data_basic_var.columns)
```
